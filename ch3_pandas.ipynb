#DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환

```python
import numpy as np

col_name1 =['col1']
list1 =[1,2,3]
array1 = np.array(list1)
print('array1 shape:',array1.shape)
#리스트를 이용해 DataFrame 생성
df_list1 = pd.DataFrame(list1,columns=col_name1)
print('1차원 리스트로 만든 DataFrame:\n',df_list1)
#넘파이 ndarray 를 이용해 DataFrame 생성
df_array1 = pd.DataFrame(array1, columns=col_name1)
print('1차원 ndarray로 만든 DataFrame:\n',df_array1)
```

#2차원 형태의 데이터로 DataFrame 만들기 =>칼럼명 갯수 주의

```python
#3개의 칼럼명이 필요
col_name2 = ['col1','col2','col3']

#2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환
list2 = [[1,2,3],[11,12,13]]
array2 = np.array(list2)
print('array2 shape:',array2.shape)
df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 list로 만든 DataFrame:\n',df_list2)

df_array2 = pd.DataFrame(array2,columns = col_name2)
print('2차원 ndarray로 만든 DataFrame:\n',df_array2)
```

# 딕셔너리를 DataFrame으로 변환

```python
#key는 문자열 칼럼명으로, value는 리스트형 칼럼 데이터로
dict = {'col1':[1,11], 'col2':[2,22], 'col3':[3,33]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n',df_dict)
```

# values:
>DataFrame을 넘파이 ndarray로 변환

```python
array3=df_dict.values
print('df_dict.values 타입:',type(array3), 'df_dict.values shape:',array3.shape)
print(array3)
```

# tolist(), to_dict() 
>DataFrame을 리스트와 딕셔너리로 변환

```python
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입:',type(list3))
print(list3)
```

```python
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입:',type(dict3))
print(dict3)
```

# DataFrame의 칼럼 데이터 세트 생성과 수정하기
>DataFrame[] 내에 새로운 칼럼명 입력 후 값 할당!

```python
titanic_df['Age_0'] = 0
titanic_df.head(3)
```

```python
titanic_df['Age_by_10'] = titanic_df['Age']* 10 #Age_by_10 이라는 칼럼 셑은 Age *10 의 값을 할당
titanic_df['Family_No'] = titanic_df['SibSp'] + titanic_df['Parch'] + 1 #Family_No 라는 칼럼 셀은 SibSp와 Parch 값을 더한 후 1을 더한 값을 할당한다
titanic_df.head(3) #위에 3줄만 출력
```

# DataFrame 내 기존 칼럼 값 일괄 업데이트 하기
> 업데이트 원하는 칼럼명 지정 후 할당

```python
titanic_df['Age_by_10'] = titanic_df['Age_by_10'] + 100 #원래의 값에 100을 더한 값으로 일괄 업데이트
titanic_df.head(3)
```

# drop() :
DataFrame 데이터 삭제
> axis=0: row 방향, axis=1: columns 방향 드롭


```python
titanic_drop_df= titanic_df.drop('Age_0',axis=1) #'Age_)' 이라는 칼럼 제거
titanic_drop_df.head(3)
```

```python
titanic_df.head(3) #그러나 원래의 변수에선 'Age_0'이라는 변수가 사라지지 않았다
```

# inplace = True
원래 값 자체에 할당하기

```python
drop_result = titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis = 1, inplace=True)
print('inplace=True 로 drop 후 반환된 값 :', drop_result)
titanic_df.head(3) #이제는 원래의 변수에서 위의 칼럼들이 모두 사라졌다
```

```python
#axis=0 으로 설정해 index0,1,2 로우를 삭제해보장
pd.set_option('display.width',1000)
pd.set_option('display.max_colwidth',15)
print('### before axis 0 drop ###')
print(titanic_df.head(3))

titanic_df.drop([0,1,2], axis=0, inplace=True) #인덱스가 0,1,2인 행들 모두 삭

print('### after axis 0 drop###')
print(titanic_df.head(3))
```

# index 객체 추출하기: .index

# reset_index():
> 새롭게 인덱스를 연속 숫자형으로 할당, 'index' 라는 새로운 칼럼명으로 추가

```python
titanic_reset_df = titanic_df.reset_index(inplace = False)
titanic_reset_df.head(3)
```

```python
print('## before reset_index ###')
value_counts = titanic_df['Pclass'].value_counts()
print(value_counts)
print('value_counts 객체 변수 타입:' , type(value_counts))

new_value_counts = value_counts.reset_index(inplace=False)
print('### After reset_index ###')
print(new_value_counts)
print('new_value_counts 객체 변수 타입:', type(new_value_counts)) #series 에 reset_index()를 적용할 경우 DataFrame이 반환된다
#value_counts 는 많이 나온 순으로 인덱스 정렬함 (내림차순)
```
