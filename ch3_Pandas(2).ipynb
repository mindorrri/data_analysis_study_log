# 결손 데이터 처리하기
Nan 여부 확인 :isna()
>Nan 을 다른 값으로 대체 : fillna()

```python
#결손 데이터 여부 확인
titanic_df.isna().head(3)
```

```python
#isna().sum() :True=1, False=0
titanic_df.isna().sum()
```

```python
#결손 데이터 대체하기
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('C000')
titanic_df.head(3)
```

```python
#'Age' 칼럼의 NAN 값을 평균 나이로, 'Embarked' 칼럼의 NAN 값을 'S' 로 대체
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna('S')
titanic_df.isna().sum()
```

# apply lambda 식으로 데이터 가공
함수의 선언과 함수 내 처리를 한줄의 식으로 쉽게 변환
> def - return

```python
def get_square(a):
    return a**2

print('3의 제곱은:',get_square(3))
```

```python
lambda_square = lambda x : x**2
print('3의 제곱은:',lambda_square(3))
```

```python
 a = [1,2,3]
squares = map(lambda x : x**2, a) #map(): list를 한번에 계산
list(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x :len(x)) #이름의 길이를 열로 따로 만들기
titanic_df[['Name','Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x >=15 else 'Adult') #'Child_Adult'라는 열 추가: 나이가 15보다 적으면 child, 아니면 Adult
titanic_df[['Age', 'Child_Adult']].head(8) #열 Age, Child_Adult 앞의 8개 행만 출력
```

# lambda 는 if, else 를 지원한다. else if 는 지원하지 않는다
if의 경우 반환값 먼저 쓴다. else의 경우 반환값을 뒤에 쓴다
lambda x : 'Child' if x<=15 else 'Adult'

```python
titanic_df ['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Age_cat'].value_counts() #항목별 갯수 세기
```

```python
#나이에 따라 세분화된 분류를 수행하는 함수 생성
def get_category(age):
    cat = ''
    if age <=15: cat = 'Baby'
    elif age <=12: cat = 'child'
    elif age <=18 : cat = 'Teenager'
    elif age <=25 : cat = 'Student'
    elif age <=35 : cat = 'Young Adult'
    elif age <=60 : cat = 'Adult'
    else: cat = 'Elderly'

    return cat

titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x)) #칼럼에 추가
                                                
titanic_df[['Age','Age_cat']].head()
```
